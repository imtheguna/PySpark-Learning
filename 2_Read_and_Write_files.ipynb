```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Read and Write Data Using PySpark") \
    .getOrCreate()
```

```python
## Reading and Writing CSV Files

## Read

df = spark.read.csv('/content/sample_data/mnist_test.csv',header=False, inferSchema=True)

df.show(1)

## Write

data = [("Alice", 34), ("Bob", 45), ("Cathy", 29)]

columns = ["Name", "Age"]

df = spark.createDataFrame(data, columns)

output_path = '/content/sample_data/users.csv'

df.write.csv(output_path, header=True, mode="overwrite")
```

```python
## Reading and Writing JSON Files

## Read

df = spark.read.json('/content/sample_data/anscombe.json')

df.show()

## Write

data = [("Alice", 34), ("Bob", 45), ("Cathy", 29)]

columns = ["Name", "Age"]

df = spark.createDataFrame(data, columns)

output_path = '/content/sample_data/users.json'

df.write.json(output_path,mode="overwrite")
```

```python
## Reading and Writing Parquet Files

## Write

data = [("Alice", 34), ("Bob", 45), ("Cathy", 29)]

columns = ["Name", "Age"]

df = spark.createDataFrame(data, columns)

output_path = '/content/sample_data/users'

df.write.parquet(output_path,mode="overwrite")

## Read

df = spark.read.parquet(output_path)

df.show()
```
